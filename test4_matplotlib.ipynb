# contours 等高线图

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def height(x,y):
    return(1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
```

```python
n=256
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
```

```python
X,Y=np.meshgrid(x,y)#transform to mesh (I`dont know this clearly actually)
plt.contourf(X,Y,height(X,Y),8,alpha=0.75,cmap=plt.cm.cool)#contour fill  parameter'8' represents 9 colors
C=plt.contour(X,Y,height(X,Y),8,colors='k',linewidths=1)
plt.clabel(C,inline=True,fontsize=10)
plt.xticks(())#clear the xticks
plt.yticks(())
plt.show()
```

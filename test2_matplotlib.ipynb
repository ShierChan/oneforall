# The fun "plt.scatter" can create

```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
n=1024 #just a constant
x=np.random.normal(0,1,n)
y=np.random.normal(0,1,n)
T=np.arctan2(x,y)#just for color value
plt.xlim((-2,2))
plt.ylim((-2,2))
print(T)
```

```python
plt.scatter(x,y,marker="o",c=T,alpha=0.5,s=50)
```
